```python
import numpy as np
import pandas as pd
students = np.random.randint(50,101,size=(50,5))
stu = pd.DataFrame(students,
             columns=['國文','英文','數學','地理','歷史'],
             index=[f'第{i}號'for i in range(1,51)]
             )
stu
```

```python
stu.index
```

```python
stu.columns
```

```python
#使用陣列運算取出所有學生的總分
sum_series = stu['國文'] + stu['英文'] + stu['數學'] + stu['地理'] + stu['歷史']
avarage_series = sum_series / 5
avarage_series
```

```python
stu['總分'] = sum_series
stu['平均'] = avarage_series
stu
```

```python
#選取分數
stu.iloc[:,:5]
stu[['國文', '英文', '數學', '地理', '歷史']]
```
