```python
import pandas as pd

df1 = pd.read_csv('opendata112N010.csv')
df2 = df1.rename(columns={'statistic_yyy':'統計年',
                    'site_id':'區域別',
                    'people_total':'年底人口數',
                    'area':'土地面積',
                    'population_density':'人口密度'})
df3 = df2.drop(index=0)
df3
```

```python
df3.iloc[-10:]
```

```python
#刪除nan
df4 = df3.dropna()
df4
```

```python
df4.iloc[-10:]
```

```python
df5 = df4.drop(columns=['統計年'])
df5
```

```python
df6 = df5.set_index('區域別')
df6
```

```python
df6.info()
#object -> 字串
```

```python
df6['土地面積'] = df6['土地面積'].astype(float)
```

```python
df6['年底人口數'] = df6['年底人口數'].astype(int)
```

```python
#不好的處理方式
nums = []
for num in df6['年底人口數']:
    try:
        nums.append(int(num))
    except:
        nums.append(0)
nums 
```

```python
def to_int(value):
    try:
        return int(value)
    except:
        return 0

df6['年底人口數'] = df6['年底人口數'].map(to_int)
df6['人口密度'] = df6['人口密度'].map(to_int)
```

```python
df6.info()
df6
```

```python
#dataFrame儲存為csv,excel,json

df6.to_csv('人口分佈.csv')
```
