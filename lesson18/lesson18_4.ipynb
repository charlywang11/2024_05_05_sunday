```python
import numpy as np
import pandas as pd
students = np.random.randint(50,101,size=(50,5))
stu = pd.DataFrame(students,
             columns=['國文','英文','數學','地理','歷史'],
             index=[f'第{i}號'for i in range(1,51)]
             )
stu
```

```python
def abc(s)->float:
    return s.mean()

stu.apply(abc)
```

```python
def efg(s)->pd.Series: #return series
    return pd.Series([s.mean(),s.max(),s.min()],
                     index=['平均',"最高分","最低分"])

stu.apply(efg,axis=1)
```
