```python
import pandas as pd

df1 = pd.read_csv('各鄉鎮市區人口密度/opendata103N010.csv')
df1.columns = df1.iloc[0]
df1 = df1[['區域別', '統計年', '年底人口數', '土地面積', '人口密度']]
df1 = df1.drop(0)
df1 = df1.dropna()

columns_to_convert = ['統計年', '年底人口數', '土地面積', '人口密度']
df1[columns_to_convert] = df1[columns_to_convert].apply(pd.to_numeric, errors='coerce').fillna(0)
#pd.to_numeric()將一列數據轉換為數字型態  #errors='coerce'遇到無法轉換的值時，將這些值轉換為NaN(空值) #.fillna()將NaN值填充為指定的值(0)

dtype_dict = {'統計年':'int',
              '區域別':'str',
              '年底人口數':'int',
              '土地面積':'float',
              '人口密度':'int'}
df1 = df1.astype(dtype_dict)
df1 = df1.set_index(keys=['區域別', '統計年'])

df1
```

```python
import pandas as pd
import os

def formatData(filename:str) -> pd.DataFrame:
    df = pd.read_csv(filename)
    df.columns = df.iloc[0]
    df = df[['區域別', '統計年', '年底人口數', '土地面積', '人口密度']]
    df = df.drop(0)
    df = df.dropna()

    columns_to_convert = ['統計年', '年底人口數', '土地面積', '人口密度']
    df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce').fillna(0)
    
    dtype_dict = {'統計年':'int',
                  '區域別':'str',
                  '年底人口數':'int',
                  '土地面積':'float',
                  '人口密度':'int'}
    df= df.astype(dtype_dict)
    df = df.set_index(keys=['區域別', '統計年'])
    
    return df


current_path = os.path.abspath(__name__)
main_path = os.path.dirname(current_path)

dataFolder_path = os.path.join(main_path,'各鄉鎮市區人口密度')
filenames = os.listdir(dataFolder_path)
abs_name = [os.path.join(dataFolder_path, name) for name in filenames if 'opendata' in name]

result = [formatData(ads_n) for ads_n in abs_name]

final_df = pd.concat(result)
final_df.sort_values(by='區域別', inplace=True)
final_df.info()
final_df
```
